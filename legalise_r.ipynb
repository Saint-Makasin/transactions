#Разделение выборки

```python
y = np.array(data.Class)
X = data.drop('Class', axis = 1)
```

#Создание и обучение модели

```python
df.Amount.plot()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, stratify=y, test_size = 0.25)
```

```python
X_train.shape
```

#Оценка модели

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
print(f'accuracy = {accuracy_score(y_test, y_pred)}')
print(f'precision = {precision_score(y_test, y_pred)}')
print(f'recall = {recall_score(y_test, y_pred)}')
print(f'f1 = {f1_score(y_test, y_pred)}')
cm = confusion_matrix(y_test, y_pred, labels=dtc.classes_) # создание массива матрицы
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                               display_labels=dtc.classes_) # создание display matrix
disp.plot() # вывод матрицы
```

```python
rfc = RandomForestClassifier(n_estimators=25, max_depth=10, max_features=5, max_samples=20000)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
print(f'accuracy = {accuracy_score(y_test, y_pred)}')
print(f'precision = {precision_score(y_test, y_pred)}')
print(f'recall = {recall_score(y_test, y_pred)}')
print(f'f1 = {f1_score(y_test, y_pred)}')
cm = confusion_matrix(y_test, y_pred, labels=dtc.classes_) # создание массива матрицы
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                               display_labels=dtc.classes_) # создание display matrix
disp.plot() # вывод матрицы
```

#Выводы

Наша модель обучена распознавать мошеннические операции, но из-за недостаточного количества данных результаты не идеальны. Модель с высокой точностью определяет обычные операции, но ошибается в четверти случаев, принимая мошеннические операции за обычные.
